```python
# Function for determining the number and percentages of missing values

def missing (tree):
    missing_number = tree.isnull().sum().sort_values(ascending=False)
    missing_percent = (tree.isnull().sum()/tree.isnull().count()).sort_values(ascending=False)
    missing_values = pd.concat([missing_number, missing_percent], axis=1, keys=['Missing_Number', 'Missing_Percent'])
    return missing_values
```

```python
def is_outlier(x, k):
    upper = np.quantile(x, 0.75)
    lower = np.quantile(x, 0.25)
    iqr = upper - lower
    max_treshold = upper + iqr * k
    min_treshold = lower - iqr * k
    if max_treshold > x.max():
        max_treshold = x.max()
    if min_treshold < x.min():
        min_treshold = x.min()
    print('-----',x.name,'-----')
    print('Upper Limit: ', max_treshold, '\n', 'Lower Limit: ', min_treshold, sep='')
    print('Number of outliers in the lower bound:', len(tree[x<min_treshold]))
    print('Number of outliers in the upper bound:', len(tree[x>max_treshold]))
```

```python
for i in tree[numeric_features]:
    is_outlier(tree[i],3)
```

### Deal with Outliers

#### *ii. Drop Outliers*

```python
def remove_outlier(tree_in, col_name,):
    q1 = tree_in[col_name].quantile(0.25)
    q3 = tree_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-3*iqr
    fence_high = q3+3*iqr
    tree_in.loc[(tree_in[col_name] < fence_low) | (tree_in[col_name] > fence_high),col_name]= np.nan
```

## 3. Feature Engineering

### My Plan of Feature Extraction
- First, I decided to produce&transform a new column with ``Horizontal_Distance_To_Hydrology`` and ``Vertical_Distance_To_Hydrology`` columns. New column will contain the values of **Hypotenuse** of ``horizantal`` and ``vertical`` distances.

- As second, we can produce&transform an additional column which contains **average** of Horizantal Distances to Hydrology and Roadways.

- Third, I decided to transform a new column which contains **average** of `Elevation` and ``Vertical_Distance_To_Hydrology`` columns. So that, there is no need to have ``Horizontal_Distance_To_Hydrology`` and ``Vertical_Distance_To_Hydrology`` columns, because I have new columns which represent more value than them. I decide to drop these columns.

- Note that, after seeing the result of the models, there may be a possibility of making minor changes to the features in the modeling phase.

```python
tree[numeric_features].T
```

```python
# 1. Produce Hypotenuse of horizantal and vertical distances.
  
Hypotenuse_Distance_Hydrology = np.hypot(tree.Horizontal_Distance_To_Hydrology, tree.Vertical_Distance_To_Hydrology)
tree["Hypotenuse_Distance_Hydrology"] = pd.DataFrame(Hypotenuse_Distance_Hydrology)
tree.head()
```

```python
# 2.  Produce&transform an additional column which contains average of Horizantal Distances to Hydrology and Roadways.

tree['Average_Distance_Hydrology_Roadways'] = tree[['Horizontal_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways']].mean(axis=1)
tree.head()
```

```python
# 3 . Produce&transform an additional column which contains average of Elevation and Vertical_Distance_To_Hydrology columns.

tree['Average_Distance_Elevation_VerHydrology'] = tree[['Vertical_Distance_To_Hydrology', 'Elevation']].mean(axis=1)
tree.head()
```

```python
# 4 . no need to have Horizontal_Distance_To_Hydrology and Vertical_Distance_To_Hydrology columns, we drop them.
    
tree.drop (['Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Hydrology'], inplace=True, axis=1)   
```

```python
# reorder columns
tree = tree.reindex(list(tree.columns[:8])+["Hypotenuse_Distance_Hydrology","Average_Distance_Hydrology_Roadways","Average_Distance_Elevation_VerHydrology"]+list(tree.columns[8:-3]), axis=1)
```
